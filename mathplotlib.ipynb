```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,5,8,1,9,2])
y=np.array([0,50,0,40,40,0])
plt.plot(x,y, marker='o')
plt.show()
x
```

```python
plt.plot(x,y,linestyle = 'dashed', marker = 'o')
# plt.plot(y[0],marker = 'o')
```

```python
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x,y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.bar(x,y,width=0.5)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
l=['a','b','c','d','e','f']
plt.pie(x,labels = l)
plt.show()
```
