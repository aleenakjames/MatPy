```python
import numpy as np
# %matplotlib inline
advert = pd.read_csv("advertising.csv")
advert.head()#only 5 rows
```

```python
advert.columns
```

```python
sb.pairplot(advert,x_vars = ['TV','Radio','Newspaper'],y_vars = 'Sales',height = 10,aspect = 1,kind = 'reg')
```

```python
x=advert[['TV']]
y=advert['Sales']
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test = train_test_split(x,y,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)
print(linreg.coeff_)
print(linreg.intercept)
```
